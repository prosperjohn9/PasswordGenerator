```python
# Obtain the website name from the user
site_name = input("Please enter the website for which the password will be generated: ")

# Find the number of characters within the website name
length = len(site_name)

'''
Check validity of length
Display error if invalid, set the password if valid

'''
if length < 5 or length > 9:
  print("Invalid website length, password cannot be generated")
else:
  # Set the password as the last 2 letters concatenated with the first 2 letters
  password = site_name[-2:] + site_name[:2]

  # Append the number of characters to the password
  password = password + str(length)

  # Append the textual version of this number and the symbol on the keyboard
  if length == 5:
    text = "five"
    symbol = "%"
  elif length == 6:
    text = "six"
    symbol = "&"
  elif length == 7:
    text = "seven"
    symbol = "/"
  elif length == 8:
    text = "eight"
    symbol = "("
  else:
    text = "nine"
    symbol = ")"

  password = password + text + symbol

  # Display the result
  print("Your password for the website ", site_name, " is ", password, ".", sep="")
```
